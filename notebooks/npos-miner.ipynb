```
  [
    "kGgmgGKJHRMLkYRnjNSmu6ntrgNnhMdizphbpJSXAKv7sm5Yf",
    "184,501,845,018,450",
    [ "kGgmgGKJHRMLkYRnjNSmu6ntrgNnhMdizphbpJSXAKv7sm5Yf" ]
  ],
```

the own stake of `kGg...5Yf` is `99,999,999,999,999,900`, why does it show `184,501,845,018,450` here?

Answer: in pallet-staking > `fn get_npos_voters`

```
let voter_weight = weight_of(&voter);
```

voter_weight is computed by `(voter's slashable balance) / (total issuance / u64::MAX).max(1)`

VARA's total issuance is 10 Bn, which is 542 times u64::MAX

```
pub const MAX: u64 = u64::MAX; // 18_446_744_073_709_551_615u64
```
https://doc.rust-lang.org/std/u64/constant.MAX.html

so the voter weight is the own stake divided by 542

```
>> (10_000_000_000_000_000_000_000u128 / u64::MAX as u128).max(1)
542
>> 99_999_999_999_999_900u128 / 542
184501845018450
```
