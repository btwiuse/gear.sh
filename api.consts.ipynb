```typescript
import { GearApi } from "https://gear-js.deno.dev/api/src/GearApi.ts?refresh"
const api = await GearApi.create({
    providerAddress: "wss://rpc.vara-network.io",
})
```

```typescript
for (let pallet of Object.keys(api.consts)) {
    for (let item of Object.keys(api.consts[pallet])) {
        console.log(`# ${pallet}.${item}`)
        let i = api.consts[pallet][item]
        console.log(i.toHuman())
        console.log()
    }
}
```
